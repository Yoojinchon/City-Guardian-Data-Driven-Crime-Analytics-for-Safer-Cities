#One column Section

```python
plt.figure(figsize=(12, 6))
top_cities = df.groupby('agency_jurisdiction')['violent_crimes'].mean().sort_values(ascending=False).head(10)
top_cities.plot(kind='barh')
plt.title('Top 10 Cities by Average Violent Crimes')
plt.xlabel('Average Violent Crimes')
plt.ylabel('City')
plt.grid(True)
plt.show()
```

```python
#Crimes
top_jurisdictions = df[df['agency_jurisdiction'] != "United States"].groupby('agency_jurisdiction')['violent_crimes'].sum().sort_values(ascending=False).head(10)
top_jurisdictions.plot(kind='barh', figsize=(15,6))
plt.title("Top 10 Jurisdictions by Violent Crimes")
plt.xlabel("Number of Violent Crimes")
plt.ylabel("Jurisdiction")
plt.show()
```

```python
#Crimes_Per_Capita
top_jurisdictions_percapita = df.groupby('agency_jurisdiction')['crimes_percapita'].mean().sort_values(ascending=False).head(10)
top_jurisdictions_percapita.plot(kind='barh', figsize=(15,6))
plt.title("Top 10 Jurisdictions by Crimes Per Capita")
plt.xlabel("Crimes Per Capita")
plt.ylabel("Jurisdiction")
plt.show()
```

Comparing the graph above, you can see that the number of crimes is higher in New York, but considering the crime rate per capita, Atlanta is higher.

#Sort

```python
#Number of crimes by year

plt.figure(figsize=(12, 6))
df.groupby('report_year')['violent_crimes', 'homicides', 'rapes', 'assaults', 'robberies'].sum().plot()
plt.title('Crimes Over the Years')
plt.xlabel('Year')
plt.ylabel('Crimes')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# New York City, NY, 2015
nyc_2015 = df[(df['agency_jurisdiction'] == 'New York City, NY') & (df['report_year'] == 2015)]

# Atlanta, GA, 2015
atlanta_2015 = df[(df['agency_jurisdiction'] == 'Atlanta, GA') & (df['report_year'] == 2015)]

crime_components = ['homicides', 'rapes', 'assaults', 'robberies']

# Visualized for New York City, NY (2015)
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
nyc_2015[crime_components].sum(axis=0).plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title("New York City, NY - Crime Composition 2015")

# Visualized for Atlanta, GA (2015)
plt.subplot(1, 2, 2)
atlanta_2015[crime_components].sum(axis=0).plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title("Atlanta, GA - Crime Composition 2015")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# New York City, NY, 1990
nyc_1990 = df[(df['agency_jurisdiction'] == 'New York City, NY') & (df['report_year'] == 1990)]

# Atlanta, GA, 1990
atlanta_1990 = df[(df['agency_jurisdiction'] == 'Atlanta, GA') & (df['report_year'] == 1990)]

crime_components = ['homicides', 'rapes', 'assaults', 'robberies']

# Visualized for New York City, NY (1990)
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
nyc_1990[crime_components].sum(axis=0).plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title("New York City, NY - Crime Composition 1990")

# Visualized for New York City, NY (1990)
plt.subplot(1, 2, 2)
atlanta_1990[crime_components].sum(axis=0).plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title("Atlanta, GA - Crime Composition 1990")

plt.tight_layout()
plt.show()
```

Let's compare 1990 with the highest crime rate. The crime composition of New York in the 1990s and the current crime composition are very different, but Atlanta is very similar to the 1990s. While New York City has been trying hard to control crime, Atlanta can speculate that is not the case.
